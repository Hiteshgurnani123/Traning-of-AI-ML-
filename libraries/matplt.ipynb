```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]

# ms --> marker size
# mfc --> marker face color
# mec --> marker edge color
# lw --> line width
# ls --> line style
# alpha --> transparency

plt.plot(x,y,marker = 'o', ms= 15, mfc = 'r', mec = 'b', color = 'black', lw = 2, ls = '--', alpha = 0.7, label = 'lineplot')
plt.title('lineplot')
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.legend()
plt.savefig('lineplot.png')
plt.show()
```

# Scatter Plot

```python
# Scatter Plot --> For Numerical Values

plt.scatter(x,y, marker = 'o', s = 120, c = 'red', edgecolor = 'black', alpha =  0.7, label = 'scatterplot')
plt.savefig('scatterplot.png')
plt.show()
```

# Bar Graph

```python
x = ['np', 'up', 'rj', 'gj']
y = [10,20,30,40]
colours = ['red', 'green', 'blue', 'yellow']
hatch_pattern = ['//', 'xx', '++', 'oo']
plt.barh(x,y,color = colours, hatch = hatch_pattern, label = x, edgecolor = 'black', alpha = 0.7)
plt.grid()
plt.legend()
plt.savefig('barhplot.png')
plt.show()
```

```python



#Random Data Generation 
data = np.random.random(1000)

#Creation of Histogram
plt.hist(data, bins =30, color = 'purple', edgecolor = 'black')

plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('frequency')
plt.savefig('histogram.png')

plt.show()
```

```python
arr = np.random.randint(20000, 40000, 100)
plt.hist(arr, bins= 20)
plt.savefig('histogram2.png')
plt.show()
```

# Pie - Chart

```python
x = ['np', 'up', 'rj', 'gj']
y = [10,20,30,40]
explodes = [0.1, 0, 0, 0]  # explode the first slice
plt.pie(y, labels = x, autopct = '%1.f%%', explode = explodes, shadow = True, startangle = 70)
plt.savefig('pie_chart.png')
plt.show()
```

# SubPlot

```python
x = np.linspace(0, 10, 100)
plt.figure(figsize=(10, 6))
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.title('Sin')
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.title('cos')
plt.subplot(2,2,3)
plt.plot(x, np.tan(x))
plt.title('tan')
plt.subplot(2,2,4)
plt.plot(x, np.log(x))
plt.title('log')
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig('subplots.png')
plt.show()



```
