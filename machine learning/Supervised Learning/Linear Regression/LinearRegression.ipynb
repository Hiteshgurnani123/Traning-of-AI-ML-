```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Used_Bikes.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
cat_col = df.select_dtypes(include='object')
cat_col.head()
```

```python
cat_col = cat_col.drop(['bike_name', 'city'], axis = 1)
cat_col.head()
```

```python
num_col = df.select_dtypes(exclude= 'object')
num_col.head()
```

```python
cat_col.head()
```

```python
pd.get_dummies(cat_col).astype(int).head()
```

```python
# label encoding
cat_col['owner'].value_counts()
```

```python
dt = {
    'First Owner':1, 
    'Second Owner':2,
    'Third Owner':3,
    'Fourth Owner Or More':4
    
    }
```

```python
cat_col['owner'] = cat_col['owner'].map(dt)
```

```python
cat_col.head()
```

```python
cat_col['brand'].value_counts()
```

```python
dt1 ={i: key for key , i in enumerate(list(cat_col['brand'].unique()))}
dt1
```

```python
cat_col['brand'] = cat_col['brand'].map(dt1)
```

```python
cat_col.head()
```

```python
final_dt = pd.concat([cat_col,num_col], axis = 1)
```

```python
final_dt.head()
```

```python
matrix = final_dt.corr()
matrix
```

```python
import seaborn as sns
```

```python
sns.heatmap(matrix, annot = True, cmap = 'coolwarm')
```

```python
x = final_dt.drop('price', axis = 1)
y = final_dt['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
df1 = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred})
df1
```

```python
from sklearn import metrics
print("mse :" , metrics.mean_squared_error(y_test, y_pred))
print("mae :", metrics.mean_absolute_error(y_test, y_pred))
print("rmse :", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))
plt.scatter(y_test, y_pred, alpha = 0.6, edgecolor = 'k')
plt.title('Actual  vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw = 3)
plt.show()
```
